```python
import numpy as np
import pandas as pd
import re

import larch
from larch.xafs import autobk, pre_edge
from larch.io import read_xdi, create_athena, read_athena, read_session


import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib.ticker import MultipleLocator
from matplotlib import colormaps

cm = 1/2.54
font_files = fm.findSystemFonts(r'C:\Users\giuli\OneDrive\Documenti\Università\Dottorato\Burocrazia\Logo UNIPV\UNIPV-Font\UNIPV - Font')
for font in font_files:
    fm.fontManager.addfont(font)

plt.rcParams['font.family'] = 'Roboto'
```

```python
fname='HEO4_Co_NR.prj'
dataCo=read_athena(fname)

groups = list(dataCo.groups.keys())
print(dataCo.groups)

### CHANGING GROUP KEYS ###
for group in groups:
    Temp = re.split('_', group)[1]
    dataCo.groups[Temp] = dataCo.groups[group]

for group in groups:
    del dataCo.groups[group]
```

```python
### MULTIPLE PLOT ###
viridis = colormaps['viridis'].resampled(12).colors
fig, [ax, ax1]=plt.subplots(2, figsize= (15*cm,15*cm))

for n, name in enumerate(dataCo.groups):
    plotdata = dataCo.groups[name]
    pre_edge(plotdata.energy, plotdata.mu, plotdata, pre2=-50, norm1= 100, norm2=700, nnorm=2)
    autobk(plotdata.energy,plotdata.mu, plotdata, kweight=2,clamp_hi=2000)

    ax.plot(plotdata.energy, plotdata.flat-0.2*n, label=name+'°C', c=viridis[n])
    ax1.plot(plotdata.k,plotdata.chi*plotdata.k**2-n*0.25,c=viridis[n])

ax.set_xlim(plotdata.e0-40, plotdata.e0+100)
ax.set_xlabel('E (eV)')
ax.set_ylabel(r'$\mu$(E)')
ax.set_yticks([])
ax.grid(axis='x')
ax.xaxis.set_minor_locator(MultipleLocator(2.5))
ax.grid(axis='x', which='minor', alpha=0.2)
ax.set_title('XANES', fontsize=10)
ax.tick_params(axis='x', labelsize=8)


ax1.grid(axis='x')
ax1.tick_params(axis='x', labelsize=8)
ax1.xaxis.set_minor_locator(MultipleLocator(.25))
ax1.grid(axis='x', which='minor', alpha=0.2)
ax1.set_xlabel(r'k ($\AA^{-1}$)')
ax1.set_ylabel(r'$k^2\chi (\AA^{-2})$')
ax1.set_yticks([])
ax1.set_xlim(2,12)
ax1.set_title('EXAFS', fontsize=10)

fig.legend(loc='lower center', ncol=6, bbox_to_anchor=(0.5,-0.05), fontsize=8)
fig.suptitle('HEO4 hydrostar', fontsize=12)
fig.tight_layout()
fig.savefig('HEO4 hydrostar XAFS.png', bbox_inches='tight', dpi=300)

plt.show()
```
